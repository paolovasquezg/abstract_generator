```python
import matplotlib.pyplot as plt
import random
import math
from samila import GenerativeImage, Projection, Marker

def f1(x, y):
    result = random.uniform(-1,1) * x**2  - math.sin(y**2) + abs(y-x)
    return result
def f2(x, y):
    result = random.uniform(-1,1) * y**3 - math.cos(x**2) + 2*x
    return result
g = GenerativeImage(f1, f2)
```

```python
from PIL import Image
from IPython.display import display
from IPython.display import clear_output
import os
import sys
from colorama import * 
import random

projections = ["Rectilinear", "Polar", "Aitoff", "Hammer", "Lambert", "Mollweide", "Random"]
markers = ["Point", "Pixel", "Circle", "Triangle_down", "Triangle_up", "Triangle_left", "Triangle_right", "Tri_down", "Tri_up", "Tri_left", "Tri_right", "Octagon", "Square", "Pentagon", "Plus", "Plus_filled", "Star", "Hexagon_vertical", "Hexagon_horizontal", "X", "X_filled", "Diamond", "Diamon_thin", "Vline", "Hline", "Random"]
color_r = ["Red", "Blue", "Black", "Yellow", "White", "Green"]
color_b = ["Red", "Blue", "Black", "Yellow", "White", "Green"]

sta = 0
en = 15
stp = 0.01

def plot_projection(a):
    if (a == 1):
        g.plot(projection=Projection.RECTILINEAR)
    elif (a == 2):
        g.plot(projection=Projection.POLAR)
    elif (a == 3):
        g.plot(projection=Projection.AITOFF)
    elif (a == 4):
        g.plot(projection=Projection.HAMMER)
    elif (a == 5):
        g.plot(projection=Projection.LAMBERT)
    elif (a == 6):
        g.plot(projection=Projection.MOLLWEIDE)
    elif (a == 7):
        g.plot(projection=Projection.RANDOM)
    else:
        g.plot(projection=Projection.RECTILINEAR)
    
def plot_markers(a):
    if (a == 1):
        g.plot(marker=Marker.POINT)
    elif (a == 2):
        g.plot(marker=Marker.PIXEL)
    elif (a == 3):
        g.plot(marker=Marker.CIRCLE)
    elif (a == 4):
        g.plot(marker=Marker.TRIANGLE_DOWN)
    elif(a == 5):
        g.plot(marker=Marker.TRIANGLE_UP)
    elif(a == 6):
        g.plot(marker=Marker.TRIANGLE_LEFT)
    elif(a == 7):
        g.plot(marker=Marker.TRIANGLE_RIGHT)
    elif(a == 8):
        g.plot(marker=Marker.TRI_DOWN)
    elif(a == 9):
        g.plot(marker=Marker.TRI_UP)
    elif(a == 10):
        g.plot(marker=Marker.TRI_LEFT)
    elif(a == 11):
        g.plot(marker=Marker.TRI_RIGHT)
    elif(a == 12):
        g.plot(marker=Marker.OCTAGON)
    elif(a == 13):
        g.plot(marker=Marker.SQUARE)
    elif(a == 14):
        g.plot(marker=Marker.PENTAGON)
    elif(a == 15):
        g.plot(marker=Marker.PLUS)
    elif(a == 16):
        g.plot(marker=Marker.PLUS_FILLED)
    elif(a == 17):
        g.plot(marker=Marker.STAR)
    elif(a == 18):
        g.plot(marker=Marker.HEXAGON_VERTICAL)
    elif(a == 19):
        g.plot(marker=Marker.X)
    elif(a == 20):
        g.plot(marker=Marker.X_FILLED)
    elif(a == 21):
        g.plot(marker=Marker.DIAMOND)
    elif(a == 22):
        g.plot(marker=Marker.DIAMON_THIN)
    elif(a == 23):
        g.plot(marker=Marker.VLINE)
    elif(a == 24):
        g.plot(marker=Marker.RANDOM)
    else:
        g.plot(marker=Marker.POINT)
    
def plot_color_fill(a):
    if (a == 1):
        g.plot(color = color_r[a-1])
    elif (a == 2):
        g.plot(color = color_r[a-1])
    elif (a == 3):
        g.plot(color = color_r[a-1])
    elif (a == 4):
        g.plot(color = color_r[a-1])
    elif(a == 5):
        g.plot(color = color_r[a-1])
    elif(a == 6):
        g.plot(color = color_r[a-1])
    else:
        g.plot(color = "black")

def plot_color_bg(a):
    if (a == 1):
        g.plot(bgcolor = color_b[a-1])
    elif (a == 2):
        g.plot(bgcolor = color_b[a-1])
    elif (a == 3):
        g.plot(bgcolor = color_b[a-1])
    elif (a == 4):
        g.plot(bgcolor = color_b[a-1])
    elif(a == 5):
        g.plot(bgcolor = color_b[a-1])
    elif(a == 6):
        g.plot(bgcolor = color_b[a-1])
    else:
        g.plot(bgcolor = "white")

def gallery(a,b):
    image_paths0 = [
        "figs/samila1.png",
        "figs/samila2.png",
        "figs/samila3.png",
        "figs/samila4.png",
        "figs/samila5.png",
        "figs/samila6.png",
        "figs/samila7.png",
        "figs/samila8.png",
        "figs/samila9.png",
        "figs/samila10.png",
        "figs/samila11.png",
        "figs/samila12.png",
        "figs/samila16.png",
        "figs/samila17.png",
        "figs/samila18.png",
    ]

    image_paths = []

    for i in range (a,b+1):
        image_paths.append(image_paths0[i])

    images = [Image.open(image_path) for image_path in image_paths]

    images_per_row = 3

    num_rows = (len(images) + images_per_row - 1) // images_per_row
    max_heights = [max(images[i].height for i in range(row * images_per_row, min((row + 1) * images_per_row, len(images)))) for row in range(num_rows)]

    total_width = sum(image.width for image in images)
    total_height = sum(max_heights)

    concatenated_image = Image.new('RGB', (total_width, total_height), (255, 255, 255))
    current_x = 0
    current_y = 0

    for i, image in enumerate(images):
        row = i // images_per_row
        concatenated_image.paste(image, (current_x, current_y))
        current_x += image.width

        if current_x >= total_width or (i + 1) % images_per_row == 0:
            current_x = 0
            current_y += max_heights[row]

    display(concatenated_image)
    
def generate_image():

    print("Enter your points range (diference between start and end below 15; step between 0,1 and 1):")
    
    print("Start:")
    sta = int(input())

    print("End:")
    en = int(input())

    print("Step")
    stp = float(input())

    clear_output(wait=True)

    if (sta > en or en - sta > 15 or stp < 0.01 or stp > 1):
        g.generate(start=0, step=0.01, stop=15)
    else:
        g.generate(start=sta, step=stp, stop=en)
    
    print("Select your projection: ")
    for i in range(len(projections)-1):
        print(str(i+1)+"." + " " + projections[i])
    option =  int(input())
    plot_projection(option)
    clear_output(wait=True)

    print("Select your maker: ")
    for i in range(len(markers)-1):
        print(str(i+1)+"." + " " + markers[i])
    option = int(input())
    plot_markers(option)
    clear_output(wait=True)

    print("Enter your marker size (between 1 and 10):")
    option = int(input())
    if (option < 1 or option > 10):
        g.plot(spot_size=1)
    else:
        g.plot(spot_size=option)
    clear_output(wait=True)

    print("Enter your image rotation:")
    option = int(input())
    g.plot(rotation=option)
    clear_output(wait=True)

    print("Select your fill color: ")
    for i in range(len(color_r)-1):
        print(str(i+1)+"." + " " + color_r[i])
    option =  int(input())
    plot_color_fill(option)
    clear_output(wait=True)

    print("Select your background color: ")
    for i in range(len(color_b)-1):
        print(str(i+1)+"." + " " + color_b[i])
    option =  int(input())
    plot_color_bg(option)
    clear_output(wait=True)

    g.seed
    plt.show()
```

```python
while (1):
    option  = 0
    while (option != 1 and option != 2 and option != 3):
        print("Abstract Image Generator")
        print("1. Generate image")
        print("2. Gallery")
        print("3. Exit")
        option = int(input())
        clear_output(wait=True)

    if (option == 1):
        print("Generating image")
        generate_image()

    elif(option == 2):
        print("Showing gallery")
        for i in range(5):
            gallery(i*3, (i*3)+2)

    else:
        print("Hope you enjoyed it!")
        break
```
