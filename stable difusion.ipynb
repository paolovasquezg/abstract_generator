# Abstract Image Generator

### Sustento de implementación

Para este generador de imagenes abstractas nos estaremos apoyando de Samila, un generador de arte generativo en Python. Básicamente crea las imagenes usando miles de puntos, cálculados por fórmulas que toman valores aleatorios. Por ello es que cada imágen es diferente a la anterior.

Dicho generador posee un conjunto de funciones que pueden recibir distintos parámetros, para así darle ciertas características generales a la imagen o creación producida. Lo que haremos es aprovechar ello para que el usuario pueda describir a alto nivel como querría que fuera su imagen, y con ello generar una imagen que se acerque a dicha descripcion.

La idea de este primer acercamiento es ver lo básico de este generador y que puede ofrecernos.

### Estructura básica

Básicamente aca importamos el generador y definimos la función que nos devuelve los valores aleatorios de los puntos.

```python
import matplotlib.pyplot as plt
import random
import math
from samila import GenerativeImage

def f1(x, y):
    result = random.uniform(-1,1) * x**2  - math.sin(y**2) + abs(y-x)
    return result
def f2(x, y):
    result = random.uniform(-1,1) * y**3 - math.cos(x**2) + 2*x
    return result
g = GenerativeImage(f1, f2)
```

### Imágen básica

```python
g.generate()
g.plot()
g.seed
plt.show()
```

### Projección

Podemos darle proyección a las imagénes

```python
from samila import Projection
g.plot(projection=Projection.RANDOM)
g.seed
plt.show()
```

### Marcador

Podemos decidir como marcar nuestra imagen

```python
from samila import Marker
g.plot(marker=Marker.DIAMOND, spot_size=10)
g.seed
plt.show()
```

### Rotación

Podemos definir que tan rotada está nuestra imágen

```python
g.generate()
g.plot(rotation=45)
```

### Rango

También podemos escoger el rango o la amplitud que tendrá nuestra imagen.

```python

g.generate(start=-2*math.pi, step=0.01, stop=0)
g.plot()
g.seed
plt.show()
```

### Color

Se puede definir tanto el color de la estructura y del fondo

```python
g.generate()
g.plot(color="red", bgcolor="white", projection=Projection.POLAR)
g.seed
plt.show()
```

Existen mas parámetros que se pueden definir, pero los anteriores son los más importantes. Estos nos permitiran construir una imagen tratando de ajustarse a lo que se describe.
